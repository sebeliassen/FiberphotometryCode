```python
def process_session(session, freq, sampling_rate=15):
    phot_df = session.df_container.data[f'photwrit_{freq}']

    comp_timestamp_idx = phot_df.columns.get_loc("LedState")
    data_columns = phot_df.columns[comp_timestamp_idx + 1:]

    for data_column in data_columns:
        brain_region, side = data_column.split(sep='_')
        data = phot_df[data_column]
        timestamp = phot_df["Timestamp"]

        guppy_df = pd.DataFrame({
            'timestamps': timestamp,
            'data': data
        })
        guppy_df.reset_index(inplace=True, drop=True)
        guppy_df["sampling_rate"] = np.nan
        guppy_df.at[0, "sampling_rate"] = sampling_rate

        trial_dirs = session.trial_dir.split('/')
        prefix = '/'.join(trial_dirs[:-1]) + '_guppy/' + trial_dirs[-1]

        if not os.path.exists(prefix):
            os.makedirs(prefix)
        
        file_name = f'{prefix}/guppy_{session.mouse_id}_{freq}_{brain_region}_{side}.csv'
        print(file_name)
        guppy_df.to_csv(file_name, index=False)
```

```python
for session in sessions:
    for freq in [415, 470]:
        process_session(session, freq)
```
